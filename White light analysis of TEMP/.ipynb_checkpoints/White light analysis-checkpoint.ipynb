```python
#Plot all columns against Time
fig, ([ax1,ax2],[ax3,ax4]) = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax1.plot(var[1].TIME, var[1].CH1, label = "No White Light")
ax1.plot(var[3].TIME, var[3].CH1, label = "OD = 2")
ax1.plot(var[5].TIME, var[5].CH1, label = "OD = 0.6")
ax1.plot(var[7].TIME, var[7].CH1, label = "OD = 0.4")
ax1.plot(var[9].TIME, var[9].CH1, label = "OD = 0.3")
ax1.plot(var[11].TIME, var[11].CH1, label = "OD = 0.2")
ax1.plot(var[13].TIME, var[13].CH1, label = "White light with no filter")
ax1.set_title('Channel 1')
ax1.set(xlabel = 'Time (s)', ylabel = 'Background noise (V)', xlim = (-3.2e-8,0))
ax1.legend(loc='upper right')

ax2.plot(var[1].TIME, var[1].CH2, label = "No White Light")
ax2.plot(var[3].TIME, var[3].CH2, label = "OD = 2")
ax2.plot(var[5].TIME, var[5].CH2, label = "OD = 0.6")
ax2.plot(var[7].TIME, var[7].CH2, label = "OD = 0.4")
ax2.plot(var[9].TIME, var[9].CH2, label = "OD = 0.3")
ax2.plot(var[11].TIME, var[11].CH2, label = "OD = 0.2")
ax2.plot(var[13].TIME, var[13].CH2, label = "White light with no filter")
ax2.set_title('Channel 2')
ax2.set(xlabel = 'Time (s)', ylabel = 'Background noise (V)', xlim = (-3.2e-8,0))
ax2.legend(loc='upper right')

ax3.plot(var[1].TIME, var[1].CH3, label = "No White Light")
ax3.plot(var[3].TIME, var[3].CH3, label = "OD = 2")
ax3.plot(var[5].TIME, var[5].CH3, label = "OD = 0.6")
ax3.plot(var[7].TIME, var[7].CH3, label = "OD = 0.4")
ax3.plot(var[9].TIME, var[9].CH3, label = "OD = 0.3")
ax3.plot(var[11].TIME, var[11].CH3, label = "OD = 0.2")
ax3.plot(var[13].TIME, var[13].CH3, label = "White light with no filter")
ax3.set_title('Channel 3')
ax3.set(xlabel = 'Time (s)', ylabel = 'Background noise (V)', xlim = (-3.2e-8,0))
ax3.legend(loc='upper right')

ax4.plot(var[1].TIME, var[1].CH4, label = "No White Light")
ax4.plot(var[3].TIME, var[3].CH4, label = "OD = 2")
ax4.plot(var[5].TIME, var[5].CH4, label = "OD = 0.6")
ax4.plot(var[7].TIME, var[7].CH4, label = "OD = 0.4")
ax4.plot(var[9].TIME, var[9].CH4, label = "OD = 0.3")
ax4.plot(var[11].TIME, var[11].CH4, label = "OD = 0.2")
ax4.plot(var[13].TIME, var[13].CH4, label = "White light with no filter")
ax4.set_title('Channel 4')
ax4.set(xlabel = 'Time (s)', ylabel = 'Background noise (V)', xlim = (-3.2e-8,0))
ax4.legend(loc='upper right')

fig.suptitle('Background response to varying white light intensity', fontsize=30)
plt.subplots_adjust(left=0, bottom=0.1, right=3, top=0.9, wspace=1, hspace=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```
