```python
#import useful files
import statistics as stat
import numpy as np
import pandas as pd
import glob
import seaborn as sns
import matplotlib.ticker as mtick
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
pulseNoAtten = pd.read_excel('Control Voltage detection/No attenuation - at pulse.xlsx')
backgroundNoAtten = pd.read_excel('Control Voltage detection/No attenuation - before pulse.xlsx')
pulseAtten = pd.read_excel('Control Voltage detection/With attenuation - at pulse.xlsx')
backgroundAtten = pd.read_excel('Control Voltage detection/With attenuation - before pulse.xlsx')
```

```python
#Splitting the DF into sensible smaller dfs for plotting purposes
r640pulseNAtt = pulseNoAtten[pulseNoAtten.Channel == 'Raman640nm']
r660pulseNAtt = pulseNoAtten[pulseNoAtten.Channel == 'Raman660nm']

r640backgroundNAtt = backgroundNoAtten[backgroundNoAtten.Channel == 'Raman640nm']
r660backgroundNAtt = backgroundNoAtten[backgroundNoAtten.Channel == 'Raman660nm']

r640pulseAtt = pulseAtten[pulseAtten.Channel == 'Raman640nm']
r660pulseAtt = pulseAtten[pulseAtten.Channel == 'Raman660nm']

r640backgroundAtt = backgroundAtten[backgroundAtten.Channel == 'Raman640nm']
r660backgroundAtt = backgroundAtten[backgroundAtten.Channel == 'Raman660nm']
```

### Raman 640nm Channel (No Attenuation) Plots

```python
xerr = [0.001,0.001,0.001,0.001,0.001]
plt.errorbar(r640pulseNAtt.ControlVoltage,r640pulseNAtt.Max,r640pulseNAtt.MaxStd, xerr, c = 'black', label = 'Not Attenuated')
plt.scatter(r640pulseNAtt.ControlVoltage,r640pulseNAtt.Max, c = 'r')

plt.errorbar(r640pulseAtt.ControlVoltage,r640pulseAtt.Max,r640pulseAtt.MaxStd, xerr, c = 'red', label = 'Attenuated')
plt.scatter(r640pulseAtt.ControlVoltage,r640pulseAtt.Max, c = 'black')

plt.yscale('log')
plt.xlim(0,0.75)
plt.ylim(1,10e4)
plt.title('Peak of Raman 640nm varying Control Voltage - log scale')
plt.xlabel('Control Voltage (V)')
plt.ylabel('Peak of Raman 640nm Channel (mV)')

plt.legend()
plt.show()
```
